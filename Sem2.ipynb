```python
def bisection(f, lhs, rhs, eps=1e-6):
    mid = (lhs+rhs)/2
    while abs(lhs-rhs) > eps:
        mid = (lhs+rhs)/2
        if f(lhs)*f(mid) > 0:
            lhs = mid
        else:
            rhs = mid
    return mid
```

```python
def func(x):
    return -12*x**4*np.sin(np.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
```

```python
def diffm(f,x,h):
    N=x.shape[0]
    d = np.zeros(N, dtype=np.float32)
    d[0] = (-3*f(x[0])+4 * f(x[1])-f(x[2]))/(2*h)
    for i in range(1,N-1):
        d[i] =(f(x[i+1])-f(x[i-1]))/(2*h)
    d[N-1]=(f(x[N-3])-4*f(x[N-2])+3*f(x[N-1]))/(2*h)
    return d
```
